```python
import numpy as np
import matplotlib.pyplot as plt

# 共有的前3个点
x_common = np.array([0, 1, 2])
y_common = np.array([0.832, 0.842, 0.844])  # 前三个点相同

# 后两个点分成6组
x_branch = np.array([4, 8])
y_branches = [
    [0.864, 0.841],  # 第一条线
    [0.888, 0.898],  # 第二条线
    [0.885, 0.872],  # 第三条线
    [0.867, 0.877],  # 第四条线
    [0.884, 0.891],  # 第五条线
    [0.889, 0.874]   # 第六条线
]
error_value = 0.003

# 结构更强的点（新的数据）
x_stronger = np.array([2, 3, 4,4, 8])  # x 位置
y_stronger = np.array([0.898, 0.918, 0.904, 0.925,0.907])  # 对应的 y 值

# 颜色设置
colors = ['purple', 'blue', 'green', 'red', 'orange', 'brown']
labels = ['Chain', 'Random', 'FullConnected', 'Layered', 'Star', 'Debate']

plt.figure(figsize=(10, 8))

# 遍历 6 条不同的分支线
for i in range(6):
    # 组合 x 和 y 数据
    x_full = np.concatenate((x_common, x_branch))
    y_full = np.concatenate((y_common, y_branches[i]))
    plt.fill_between(x_full, y_full - error_value, y_full + error_value, color=colors[i], alpha=0.3)  # 固定误差区间

    # 画折线
    plt.plot(x_full, y_full, color=colors[i], label=labels[i], marker='o', markersize=5, linestyle='-', linewidth=2)
    plt.scatter(x_full, y_full, color=colors[i], edgecolor='black', zorder=3)  # 标注点

# 标注“结构更强”的点
plt.scatter(x_stronger, y_stronger, color='black', s=100, zorder=5, marker='X', label='Stronger Structure')  # 用 X 形标记
for i in range(len(x_stronger)):
    plt.text(x_stronger[i], y_stronger[i] + 0.005, f'({x_stronger[i]}, {y_stronger[i]:.3f})', ha='center', fontsize=10)

# 设置标题和标签
plt.title("Multiagent", fontsize=14, fontweight='bold')
plt.xlabel("Scale", fontsize=12)
plt.ylabel("Quality", fontsize=12)
plt.xticks([0, 1, 2, 3, 4, 8])

# 美化网格和图例
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据：分数，类别（可以是时间或其他），以及 token 数量
scores = np.array([58, 62, 65, 70, 72, 55])
categories = np.array([1, 2, 3, 4, 5, 6])  # 比如类别或时间
tokens = np.array([100, 200, 150, 300, 250, 400])  # token 数量

plt.figure(figsize=(8, 5))

# 绘制气泡图
plt.scatter(scores, categories, s=tokens, color='blue', alpha=0.6, edgecolors="w", linewidth=2)

# 设置标题和标签
plt.title("Scores vs Categories with Token Count", fontsize=14, fontweight='bold')
plt.xlabel("Score", fontsize=12)
plt.ylabel("Category", fontsize=12)

# 添加标注
for i in range(len(scores)):
    plt.text(scores[i], categories[i], f'{tokens[i]} tokens', fontsize=10, ha='right')

# 美化网格
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
